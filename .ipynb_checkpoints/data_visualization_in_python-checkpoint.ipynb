```python
# let'S INTERACT with this dataset
import altair as alt
from vega_datasets import data

source = data.cars()

brush = alt.selection_interval()

points = alt.Chart(source).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color=alt.condition(brush, 'Origin', alt.value('lightgray'))
).add_params(
    brush
)

bars = alt.Chart(source).mark_bar().encode(
    y='Origin',
    color='Origin',
    x='count(Origin)'
).transform_filter(
    brush
)

points & bars
```

```python
from vega_datasets import data

df = data.seattle_weather()
df.head()
```

```python
import altair as alt

alt.Chart(df).mark_tick().encode(
    x='precipitation',
)
```

```python
alt.Chart(df).mark_bar().encode(
    alt.X('precipitation').bin(),
    y='count()'
)
```

```python
alt.Chart(df).mark_line().encode(
    x='month(date):T',
    y='average(precipitation)'
)
```

```python
alt.Chart(df).mark_line().encode(
    x='year(date):T',
    y='mean(temp_max)',
)
```

```python
alt.Chart(df).mark_bar().encode(
    x='mean(temp_max)',
    y='year(date):O'
)
```

```python
alt.Chart(df).mark_bar().encode(
    x='mean(temp_range):Q',
    y='year(date):O'
).transform_calculate(
    temp_range="datum.temp_max - datum.temp_min"
)
```

```python
alt.Chart(df).mark_bar().encode(
    x='month(date):N',
    y='count()',
    color='weather',
)
```

```python
scale = alt.Scale(domain=['sun', 'fog', 'drizzle', 'rain', 'snow'],
                  range=['#e7ba52', '#c7c7c7', '#aec7e8', '#1f77b4', '#9467bd'])
```

```python
alt.Chart(df).mark_bar().encode(
    x=alt.X('month(date):N').title('Month of the year'),
    y='count()',
    color=alt.Color('weather', legend=alt.Legend(title='Weather type'), scale=scale),
)
```

```python
alt.Chart(df).mark_point().encode(
    alt.X('temp_max').title('Maximum Daily Temperature (C)'),
    alt.Y('temp_range:Q').title('Daily Temperature Range (C)'),
    alt.Color('weather').scale(scale),
    alt.Size('precipitation').scale(range=[1, 200])
).transform_calculate(
    "temp_range", "datum.temp_max - datum.temp_min"
).properties(
    width=600,
    height=400
).interactive()
```

```python
alt.Chart(df).mark_bar().encode(
    x='count()',
    y='weather:N',
    color=alt.Color('weather:N').scale(scale),
)
```

```python
import altair as alt
from vega_datasets import data

counties = alt.topo_feature(data.us_10m.url, 'counties')
source = data.unemployment.url

alt.Chart(counties).mark_geoshape().encode(
    color='rate:Q'
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(source, 'id', ['rate'])
).project(
    type='albersUsa'
).properties(
    width=500,
    height=300
)
```

```python
import altair as alt
from vega_datasets import data

source = data.cars()

# Brush for selection
brush = alt.selection_interval()

# Scatter Plot
points = alt.Chart(source).mark_point().encode(
    x='Horsepower:Q',
    y='Miles_per_Gallon:Q',
    color=alt.condition(brush, alt.value('steelblue'), alt.value('grey'))
).add_params(brush)

# Base chart for data tables
ranked_text = alt.Chart(source).mark_text(align='right').encode(
    y=alt.Y('row_number:O').axis(None)
).transform_filter(
    brush
).transform_window(
    row_number='row_number()'
).transform_filter(
    'datum.row_number < 15'
)

# Data Tables
horsepower = ranked_text.encode(text='Horsepower:N').properties(
    title=alt.TitleParams(text='Horsepower', align='right')
)
mpg = ranked_text.encode(text='Miles_per_Gallon:N').properties(
    title=alt.TitleParams(text='MPG', align='right')
)
origin = ranked_text.encode(text='Origin:N').properties(
    title=alt.TitleParams(text='Origin', align='right')
)
text = alt.hconcat(horsepower, mpg, origin) # Combine data tables

# Build chart
alt.hconcat(
    points,
    text
).resolve_legend(
    color="independent"
).configure_view(
    stroke=None
)
```

The altair is build using the pandas dataframe. Let's begin by coding using the help of an example-


```python
import pandas as pd
data = pd.DataFrame({'a': list('CCCDDDEEE'),
                     'b': [2, 7, 4, 1, 2, 6, 8, 4, 7]})
```

The fundamanetal object in in Altair is Chart which takes the dataframe as a single argument. 


```python
import altair as alt
chart = alt.Chart(data)
```

####  encoding and marks
With chart object in hand, we can play with another attributes of it using another methods. 


let's show data as a 'one point'


```python
alt.Chart(data).mark_point()
```

#### let's visually seperate the points and mark them on the x-axis position of the points. This can be done with chart.encode() method



```python
alt.Chart(data).mark_point().encode(x='a')
```

#### let's add y-channel to the same data frame
This will plot the values corresponding to the x and y axis.

so, dictionary has been created in the form of a bar graph plotted with points. 


```python
alt.Chart(data).mark_point().encode(x='a', y= 'b')
```

#### Aggregation
now, let's take the average values of the data to have more clarity. 


```python
alt.Chart(data).mark_point().encode(x='a', y= 'average(b)')
```

#### replacing point_markings  with bar_markings
As we know that the average data is well represented in the form of bar graphs, so lets replace point-markings with bar-markings
to present the data in the form of bar graphs-


```python
alt.Chart(data).mark_bar().encode(x='a', y= 'average(b)')
```

### let's swap the bar graph horizontally by changing axis order

```python
alt.Chart(data).mark_bar().encode(y='a', x= 'average(b)')

```

#### let's now call the above variable, 'test' and examine the .json file of it. Amazing! isn't it?

```python
test = alt.Chart(data).mark_bar().encode(y='a', x='average(b)')

print (test.to_json())
```

### beautiful, isn't it! now, just change the title of both the axis and change the color of the bar_graph


```python
alt.Chart(data).mark_bar(color='firebrick').encode(
    alt.Y('a', type='nominal'),
    alt.X('b', type='quantitative', aggregate='average')
)
```

```python
alt.Chart(data).mark_bar(color='firebrick').encode(
    alt.Y('a'),
    alt.X('average(b)')
)
```

### let's pulish the visualization using chart.save() method into an .html file



```python
abc= alt.Chart(data).mark_bar().encode(
x='a',
y='average(b)')
abc.save('abc.html')
```
