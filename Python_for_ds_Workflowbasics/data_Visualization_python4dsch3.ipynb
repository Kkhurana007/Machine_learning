the questions that arise in our brains motivate us to use the knowledge of data science to testify response-
 Do cars with big engines use more fuel than cars with small engines? What does the relationship between engine size and fuel efficiency look like? Is it positive? Negative? Linear? Nonlinear? See. they can be dealt with mpg dataframe found in ggplot2

```python
url = "https://github.com/byuidatascience/data4python4ds/raw/master/data-raw/mpg/mpg.csv"
mpg = pd.read_csv(url)
```

### variables in mpg dataset

1. displ, a car’s engine size, in litres.

2. hwy, a car’s fuel efficiency on the highway, in miles per gallon (mpg). 


now, we keep displ on x-axis and hwy on y-axix


```python

chart = (alt.Chart(mpg)
  .encode(
    x='displ', 
    y='hwy')
  .mark_circle()
)
chart.mark_point() # data gets represented in the form of points

```

```python
chart.mark_circle() # representation happens in the form of circles
```

```python
# answering question number 4 based on text!

chart = (alt.Chart(mpg)
  .encode(
    x='cyl', 
    y='hwy')
  .mark_point()
)

chart.mark_point()
```

### Aesthetic mappings 
if the points, fall outside the linear line, we can change the color of the points to see the trends

This can be done by adding a third variable, like class, to a two dimensional scatterplot by mapping it to an encoding. Trying the code-


```python
chart = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy",
    color = "class", 
    )
  .mark_circle())


chart.mark_circle()
```

### let's use class for encoding the size in the same way!

```python
# second


chart = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy",
    size = "class", 
    )
  .mark_circle())


chart.mark_circle()
```

```python
# third variation (the figures are inbuild); maximum of 8 shapes to be used for one chart


chart = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy",
    shape = "class", 
    )
  .mark_point())


chart.mark_point()
```

```python
# fourth varaition (for a specific color, blue for example)
chart = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy",
    color = alt.value("blue"),  #cusotmizing colors- 
      #https://altair-viz.github.io/user_guide/customization.html#color-schemes
    )
  .mark_circle())


chart.mark_circle()
```

### Facets (diving the plots into sub-sets of data)

```python
chart_f = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy",
   )
  .mark_circle()
  .facet(
    facet = "class",
    columns = 4
  ))

chart_f.show()
```

### Facet being the combination of two variables (column, row)

```python
chart_f2 = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy",
    )
  .mark_circle()
  .facet(
    column = "drv",
    row = "cyl")
           )
  
chart_f2.show()
#> WARN row encoding should be discrete (ordinal / nominal / binned).
```

```python
# example code

alt.Chart(mpg).encode(
    x = "displ",
    y = "hwy")
    .mark_circle()
  .facet(column = "drv"))

(alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy"
    )
  .mark_circle()
  .facet(row = "cyl"))
```

```python
chartp = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy"
    )
  .mark_circle())

chartf = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy"
    )
  .transform_loess("displ", "hwy")
  .mark_line()
  )

chartp.show()  # plot1
```

```python
chartf.show()  # plot 2
```

 ### theory
 
  bar charts use bar marks, line charts use line marks, boxplots use boxplot marks, and so on. Scatterplots break the trend; they use the point mark.
  
  
 The first plot uses the 'point mark', and the second plot uses the "line mark", a smooth line fitted to the data is calculated using a transformation

```python
chartl = (alt.Chart(mpg)
  .transform_loess("displ", "hwy", groupby = ["drv"])
  .encode(
    x = "displ",
    y = "hwy",
    strokeDash = "drv"
    )
  .mark_line())
  
chartl.show()
```

```python
chartleft = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy",
    )
  .transform_loess("displ", "hwy")
  .mark_line())

chartmiddle = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy",
    detail = "drv"
    )
  .transform_loess("displ", "hwy", groupby = ["drv"])
  .mark_line())

chartright = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy",
    color=alt.Color("drv")
    )
  .transform_loess("displ", "hwy", groupby = ["drv"])
  .mark_line())

#chartleft.properties()
#chartmiddle.properties()
chartright.properties()
```

### installing multiple charts on the same plot

```python
chartp = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy"
  )
  .mark_circle()
)

chart = chartp + chartleft  

chart.show()
```

```python
base = (alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy"
    ))

chart = base.mark_circle() + base.transform_loess("displ", "hwy").mark_line()  #to verify that we use same variables in both 
# plots

chart.show()
```

### using customization variables, we may change the color of the charts https://altair-viz.github.io/user_guide/customization.html 

changing the color of the plot_with_circle

```python
base =(alt.Chart(mpg)
  .encode(
    x = "displ",
    y = "hwy"
  ))

chart = base.encode(color = "drv").mark_circle() + base.transform_loess("displ", "hwy").mark_line()
```

```python
base = (alt.Chart(mpg.rename(columns = {"class": "class1"}))
  .encode(
    x = "displ",
    y = "hwy"
    ))

chart_smooth_sub = (base
  .transform_filter(alt.datum.class1 == "subcompact")
  .transform_loess("displ", "hwy")
  .mark_line()
)  

chart = base.encode(color = "class1").mark_circle() + chart_smooth_sub

chart.properties()
```
